## Classification model in python

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

```python
#Read data

df = pd.read_csv("breast_cancer_data.csv")
```

```python
df
```

```python
print((df["diagnosis"] == "M").sum())
print((df["diagnosis"] == "B").sum())
```

```python
meancols =  [colname for colname in df.columns if "mean" in colname]
worstcols = [colname for colname in df.columns if "worst" in colname]
```

```python
df_mean = df[["diagnosis"] + meancols]
df_worst = df[["diagnosis"] + worstcols]
```

```python
#Plot data
seaborn.pairplot(df_mean)
```

```python
df_mean.corr()
```

```python
## check is there is missing data
df.isna().any()
```

```python
df.boxplot("radius_mean", by = "diagnosis")
```

```python
## Scaling
## for "mean" data
from sklearn import preprocessing

#To a range
min_max_scaler = preprocessing.MinMaxScaler()
min_max_scaler.fit_transform(df[meancols])

## Standardization
preprocessing.scale(df[meancols])
df_mean.loc[:, meancols] = preprocessing.scale(df[meancols])
df_worst.loc[:, worstcols] = preprocessing.scale(df[worstcols])
```

```python
df_worst
```

```python
## Create training and test data
from sklearn import model_selection
rownum = df.shape[0]
train_index, test_index = model_selection.train_test_split(range(rownum), train_size = 0.7, )

from sklearn import linear_model
## for "mean"
model_mean = linear_model.LogisticRegression()
#LogisticRegressionCV()
model_mean.fit(df_mean.loc[train_index, meancols], df_mean.loc[train_index, "diagnosis"])

# get importance
importance = model_mean.coef_[0]
# summarize feature importance
for i,v in enumerate(importance):
    print('Feature: %s, Score: %.5f' % (meancols[i], v))
# plot feature importance
plt.bar([x for x in range(len(importance))], importance)
plt.show()

## for "worst"
model_worst = linear_model.LogisticRegression()
model_worst.fit(df_worst.loc[train_index, worstcols], df_mean.loc[train_index, "diagnosis"])

# get importance
importance = model_worst.coef_[0]
# summarize feature importance
for i,v in enumerate(importance):
    print('Feature: %s, Score: %.5f' % (worstcols[i], v))
# plot feature importance
plt.bar([x for x in range(len(importance))], importance)
plt.show()
```

```python
#For R-like statistical analysis, see statsmodels
import statsmodels.api as sm
logit_model=sm.Logit(np.asarray(df_mean.loc[train_index, "diagnosis"] == "M"), np.asarray(df_mean.loc[train_index, meancols]))
result=logit_model.fit()
print(result.summary())
```

```python
y_pred_mean = model_mean.predict(df_mean.loc[test_index, meancols])
y_pred_worst = model_worst.predict(df_worst.loc[test_index, worstcols])

from sklearn import metrics
print(metrics.confusion_matrix(df_mean.loc[test_index, "diagnosis"], y_pred_mean))
print(metrics.confusion_matrix(df_worst.loc[test_index, "diagnosis"], y_pred_worst))
```

```python
## Calculate AUROC for "mean" data
y_train_mean_score = model_mean.decision_function(df_mean.loc[train_index, meancols])
y_pred_mean_score = model_mean.decision_function(df_mean.loc[test_index, meancols])
print(metrics.roc_auc_score(df_mean.loc[train_index, "diagnosis"], y_train_mean_score))
print(metrics.roc_auc_score(df_mean.loc[test_index, "diagnosis"], y_pred_mean_score))

## calculate for "worst" data
y_train_worst_score = model_worst.decision_function(df_worst.loc[train_index, worstcols])
y_pred_worst_score = model_worst.decision_function(df_worst.loc[test_index, worstcols])
print(metrics.roc_auc_score(df_worst.loc[train_index, "diagnosis"], y_train_worst_score))
print(metrics.roc_auc_score(df_worst.loc[test_index, "diagnosis"], y_pred_worst_score))
```

```python
## AUROC curve
fpr_mean_train, tpr_mean_train, thresholds_mean_train = metrics.roc_curve(
            df_mean.loc[train_index, "diagnosis"], y_train_mean_score, pos_label="M")
fpr_mean, tpr_mean, thresholds_mean = metrics.roc_curve(
            df_mean.loc[test_index, "diagnosis"], y_pred_mean_score, pos_label="M")
fpr_worst_train, tpr_worst_train, thresholds_worst_train = metrics.roc_curve(
            df_worst.loc[train_index, "diagnosis"], y_train_worst_score, pos_label="M")
fpr_worst, tpr_worst, thresholds_worst = metrics.roc_curve(
            df_worst.loc[test_index, "diagnosis"], y_pred_worst_score, pos_label="M")
plt.plot(fpr_mean_train, tpr_mean_train, label = "Mean train")
plt.plot(fpr_worst_train, tpr_worst_train, label = "Mean test")
plt.plot(fpr_mean, tpr_mean, label = "Worst train")
plt.plot(fpr_worst, tpr_worst, label = "Worst test")
plt.xlabel("False positive rate")
plt.ylabel("True positive rate")
plt.legend()
plt.show()
```
